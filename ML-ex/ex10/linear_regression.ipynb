```python
import torch
import numpy as np
import matplotlib.pyplot as plt
rng = np.random
```

```python
learning_rate = 0.01
training_step = 1000
display_step = 50
```

```python
X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,7.042,10.791,5.313,7.997,5.654,9.27,3.1])
Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,2.827,3.465,1.65,2.904,2.42,2.94,1.3])
```

```python
inputs = torch.from_numpy(X)
targets = torch.from_numpy(Y)
```

```python
plt.plot(X, Y, 'ro', label = 'Original data')
plt.show()
```

```python
w = torch.randn(1, requires_grad = True)
b = torch.randn(1, requires_grad = True)
```

```python
def model(x):
    return x * w + b
```

```python
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
losses = []
step = []
```

```python
for i in range(training_step):
    step.append(i)
    preds = model(inputs)
    loss = mse(preds, targets)
    losses.append(loss.detach())
    loss.backward()
    with torch.no_grad():
        w -= w.grad * learning_rate
        b -= b.grad * learning_rate
        w.grad.zero_()
        b.grad.zero_()
```

```python
plt.plot(step, losses, label = 'Original Data')
plt.legend()
plt.show()
```

```python
plt.plot(X, Y, 'ro', label='Original data')
plt.plot(X, (w * inputs + b).detach().numpy(), label='Fitted line')
plt.legend()
plt.show()
```
