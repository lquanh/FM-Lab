```python
y = loan_data['Personal Loan']
y[y != 0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def cal(x):
    knn_classifier = KNeighborsClassifier(n_neighbors = x, metric = 'minkowski', p = 2, weights = 'distance')
    knn_classifier.fit(X_train, y_train)
    y_pred = knn_classifier.predict(X_test)
    
    return accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred)
```

```python
n_neighbors = [i for i in range(1, 31)]

acc = []
pre = []
rec = []

for i in n_neighbors:
    x, y, z = cal(i)
    acc.append(x)
    pre.append(y)
    rec.append(z)

fig, axs = plt.subplots(1, 3, figsize = (20, 4))
axs[0].plot(n_neighbors, acc, marker= ".")
axs[0].set_xlabel("n_neighbors")
axs[0].set_ylabel("Accuracy")
axs[0].set_title("Ảnh hưởng của số lượng láng giềng")
axs[1].plot(n_neighbors, pre, marker= ".")
axs[1].set_xlabel("n_neighbors")
axs[1].set_ylabel("Precision")
axs[1].set_title("Ảnh hưởng của số lượng láng giềng")
axs[2].plot(n_neighbors, rec, marker= ".")
axs[2].set_xlabel("n_neighbors")
axs[2].set_ylabel("Recall")
axs[2].set_title("Ảnh hưởng của số lượng láng giềng")
```

```python
acc = []
pre = []
cal = []

keys = ['None', 'ID', 'Age', 'Experience', 'Income', 'ZIP Code', 'Family', 'CCAvg', 'Education', 'Mortgage',
        'Personal Loan', 'Securities Account', 'CD Account', 'Online', 'CreditCard']

for i in keys:
    if(i == 'None'):
        X = loan_data.drop(['Personal Loan'], axis = 1)
    else:
        X = loan_data.drop(['Personal Loan', i], axis = 1)
    X = (X - X.min()) / (X.max() - X.min())
    y = loan_data['Personal Loan']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
    knn_classifier = KNeighborsClassifier(n_neighbors = 2, metric = 'minkowski', p = 2, weights = 'distance')
    knn_classifier.fit(X_train, y_train)
    y_pred = knn_classifier.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred))
    pre.append(precision_score(y_test, y_pred))
    cal.append(recall_score(y_test, y_pred))
```

```python
fig, axs = plt.subplots(3, 1, figsize = (24, 20))
axs[0].plot(keys, acc, marker= ".")
axs[0].set_xlabel("keys")
axs[0].set_ylabel("Accuracy")
axs[0].set_title("Ảnh hưởng của từng thuộc tính")
axs[0].plot(keys, [acc[0] for _ in range(len(acc))], marker = '', label = 'Baseline')
axs[1].plot(keys, pre, marker= ".")
axs[1].set_xlabel("keys")
axs[1].set_ylabel("Precision")
axs[1].set_title("Ảnh hưởng của từng thuộc tính")
axs[1].plot(keys, [pre[0] for _ in range(len(pre))], marker = '', label = 'Baseline')
axs[2].plot(keys, cal, marker= ".")
axs[2].set_xlabel("keys")
axs[2].set_ylabel("Recall")
axs[2].set_title("Ảnh hưởng của từng thuộc tính")
axs[2].plot(keys, [cal[0] for _ in range(len(cal))], marker = '', label = 'Baseline')
```

# Dựa vào 3 biểu đồ, ta thấy những thuộc tính khi bị loại bỏ sẽ ảnh hưởng tiêu cực đến kết quả là Income, Family, CCAvg, Education, CD Account. Đây là những thuộc tính nên sử dụng để đạt được những kết quả tốt nhất.

```python
acc = []
pre = []
cal = []
X = loan_data.drop(['Personal Loan'], axis = 1)
y = loan_data['Personal Loan']
keys = ['minkowski', 'manhattan', 'euclidean', 'chebyshev']
for i in keys:
    knn_classifier = KNeighborsClassifier(n_neighbors = 5, metric = i, p = 5, weights = 'distance')
    knn_classifier.fit(X_train, y_train)
    y_pred = knn_classifier.predict(X_test)
    acc.append(accuracy_score(y_test, y_pred))
    pre.append(precision_score(y_test, y_pred))
    cal.append(recall_score(y_test, y_pred))

fig, axs = plt.subplots(1, 3, figsize = (20, 4))
axs[0].plot(keys, acc, marker= ".")
axs[0].set_ylabel("Accuracy")
axs[0].set_title("Sự khác nhau giữa các hàm tính")
axs[1].plot(keys, pre, marker= ".")
axs[1].set_ylabel("Precision")
axs[1].set_title("Sự khác nhau giữa các hàm tính")
axs[2].plot(keys, cal, marker= ".")
axs[2].set_ylabel("Recall")
axs[2].set_title("Sự khác nhau giữa các hàm tính")
```
