```python
import numpy as np
import matplotlib.pyplot as plt
from pyvi import ViTokenizer
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.datasets import load_files
from sklearn.metrics import accuracy_score, f1_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split, GridSearchCV, learning_curve
```

```python
data_train = load_files(container_path = "data/news_1135/", encoding = "utf-8")
print(len(data_train.filenames))
```

```python
with open("data/vietnamese-stopwords.txt", encoding = "utf-8") as f:
    stopwords = f.readlines()
stopwords = [x.strip().replace(" ", "_") for x in stopwords]
print(stopwords[:10])
```

```python
module_count_vector = CountVectorizer(stop_words = stopwords)
model_rf_preprocess = Pipeline([('vect', module_count_vector), ('tfidf', TfidfTransformer())])
data = model_rf_preprocess.fit_transform(data_train.data, data_train.target)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data_train.target, test_size = 0.2, random_state = 12)
```

```python
model = SVC(kernel = "linear", C = 1)
model.fit(X_train, y_train)
```

```python
model_train_pred = model.predict(X_train)
model_test_pred = model.predict(X_test)
print("Train score: {}".format(accuracy_score(y_train, model_train_pred)))
print("Test score: {}".format(accuracy_score(y_test, model_test_pred)))
```

```python
model2 = SVC(kernel = "rbf", C = 1)
model2.fit(X_train, y_train)
```

```python
model2_train_pred = model2.predict(X_train)
model2_test_pred = model2.predict(X_test)
print("Train score: {}".format(accuracy_score(y_train, model2_train_pred)))
print("Test score: {}".format(accuracy_score(y_test, model2_test_pred)))
```

```python
def plot_learning_curve(estimator, title, label_curve, X, y, ylim = None, cv = None, n_jobs = 1, train_sizes = np.linspace(.1, 1.0, 5), new_plot = False, idx_color = 0):
    if(new_plot):
        plt.title(title)
        plt.xlabel("Training examples")
        plt.ylabel("Score")
        plt.grid()
    if ylim != None:
        plt.ylim(*ylim)
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv = cv, n_jobs = n_jobs, train_sizes = train_sizes, scoring = "accuracy")
    train_scores_mean = np.mean(train_scores, axis = 1)
    train_scores_std = np.std(train_scores, axis = 1)
    test_scores_mean = np.mean(test_scores, axis = 1)
    test_scores_std = np.std(test_scores, axis = 1)

    plt.plot(train_sizes, test_scores_mean, 'o-', color = 'r', label = label_curve['test'])
    plt.plot(train_sizes, train_scores_mean, 'o-', color = 'g', label = label_curve['train'])

    plt.legend(loc="best")
    return plt
```

```python
list_C = [0.001, 0.01, 0.1, 1, 5.0, 10.0, 25.0, 50.0, 100.0]
title = ""
plt.figure(figsize=(16, 9))
for i, c in enumerate(list_C):
    plt.subplot(3, 3, i + 1)
    plt = plot_learning_curve(SVC(kernel = 'linear', C = c), title, {'test': f'CV (C={c})', 'train': f'Train (C={c})'}, X_train, y_train, (0.0, 1.01), cv = 10, n_jobs = -1, idx_color = i, new_plot = True)
plt.show()
```

```python
params_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'gamma': [0.0001, 0.001, 0.01, 0.1],
    'kernel':['linear','rbf', 'poly']
}
grid = GridSearchCV(estimator = SVC(), param_grid = params_grid, cv = 4, n_jobs = -1, scoring = "accuracy")
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
grid_pred = grid.predict(X_test)
print(accuracy_score(y_test, grid_pred))
```
