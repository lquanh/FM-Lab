```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
data = pd.read_csv("clean_data.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
X = data.drop("price", axis = 1)
y = data['price']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 51)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
lr_lasso = Lasso()
lr_ridge = Ridge()
```

```python
def rmse(yt, yp):
    return np.sqrt(mean_squared_error(yt, yp))
```

```python
lr.fit(X_train, y_train)
```

```python
lr_score = lr.score(X_test, y_test)
y_pred = lr.predict(X_test)
lr_rmse = rmse(y_test, y_pred)
lr_score, float(lr_rmse)
```

```python
lr_lasso.fit(X_train, y_train)
```

```python
lr_lasso_score = lr_lasso.score(X_test, y_test)
y_lasso_pred = lr_lasso.predict(X_test)
lr_lasso_rmse = rmse(y_test, y_lasso_pred)
lr_lasso_score, float(lr_lasso_rmse)
```

```python
lr_ridge.fit(X_train, y_train)
```

```python
lr_ridge_score = lr_ridge.score(X_test, y_test)
y_ridge_pred = lr_ridge.predict(X_test)
lr_ridge_rmse = rmse(y_test, y_ridge_pred)
lr_ridge_score, float(lr_ridge_rmse)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
```

```python
svr.fit(X_train, y_train)
svr_score = svr.score(X_test, y_test)
svr_pred = svr.predict(X_test)
svr_rmse = rmse(y_test, svr_pred)

print(svr_score, svr_rmse, y_test.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
rfr_score = rfr.score(X_test, y_test)
rfr_pred = rfr.predict(X_test)
rfr_rmse = rmse(y_test, rfr_pred)
rfr_score, float(rfr_rmse)
```

```python
y_mean = y_test.mean()
lr_ratio = lr_rmse / y_mean
lasso_ratio = lr_lasso_rmse / y_mean
ridge_ratio = lr_ridge_rmse / y_mean
svr_ratio = svr_rmse / y_mean
rfr_ratio = rfr_rmse / y_mean
model_type = ['Linear Regression', 'Lasso Regression', 'Ridge Regression', 'Support Vector Regression', 'Random Forest Regressor']
value = [lr_ratio, lasso_ratio, ridge_ratio, svr_ratio, rfr_ratio]
plt.figure(figsize = (7, 5))
colors = np.random.rand(len(model_type), 3)
bars = plt.bar(model_type, value, color = colors)
plt.ylim(min(value) * 0.6, max(value) * 1.4)
plt.title('Result')
plt.show()
```

```python
def predict_house_price(model, bath, balcony, total_sqft_int, bhk, price_per_sqft, area_type, availability, location):
    x = np.zeros(len(X.columns))
    x[0] = bath
    x[1] = balcony
    x[2] = total_sqft_int
    x[3] = bhk
    x[4] = price_per_sqft
    x = sc.transform([x])[0]
    return model.predict([x])[0]
```

```python
def get(models):
    lr_test = predict_house_price(model = models, bath = 3, balcony = 2, total_sqft_int = 1672, bhk = 3, price_per_sqft = 8971.291866, area_type = "Plot Area", availability = "Ready To Move", location = "Devarabeesana Halli")
    return lr_test
print('Linear Regression', get(lr))
print('Lasso', get(lr_lasso))
print('Ridge', get(lr_ridge))
print('SVR', get(svr))
print('RFR', get(rfr))
```

```python
import joblib
joblib.dump(rfr, 'bangalore_house_price_prediction_model.pkl')
```
