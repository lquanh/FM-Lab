```python
import os
import numpy as np
import matplotlib.pyplot as plt
from pyvi import ViTokenizer
from tqdm import tqdm
from sklearn.datasets import load_files
from sklearn.model_selection import learning_curve, train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
```

```python
INPUT = 'data/news_vnexpress'
os.makedirs("images",exist_ok=True)
```

```python
print('Các nhãn và số văn bản tương ứng trong dữ liệu')
print('----------------------------------------------')
n = 0
for label in os.listdir(INPUT):
    print(f'{label}: {len(os.listdir(os.path.join(INPUT, label)))}')
    n += len(os.listdir(os.path.join(INPUT, label)))
print('-------------------------')
print(f"Tổng số văn bản: {n}")
```

```python
data_train = load_files(container_path = INPUT, encoding = 'utf-8')
```

```python
for i in range(len(data_train.target_names)):
    print(f'{data_train.target_names[i]}- {i}')
print('--------------------------')
print(data_train.filenames[0:1])
print(data_train.target[0:1])
print(data_train.data[0:1])
print("\nTổng số văn bản: {}" .format( len(data_train.filenames)))
```

```python
with open("data/vietnamese-stopwords.txt", encoding = 'utf-8') as f:
    stopwords = f.readlines()
stopwords = [x.strip().replace(" ", "_") for x in stopwords]
print(stopwords[:10])
```

```python
module_count_vector = CountVectorizer(stop_words = stopwords)
model_rf_preprocess = Pipeline([('vect', module_count_vector), ('tfidf', TfidfTransformer())])
data = model_rf_preprocess.fit_transform(data_train.data, data_train.target)
print(len(module_count_vector.vocabulary_))
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data_train.target, test_size = 0.2, random_state = 12)
```

```python
def cross_validation(estimator):
    _, train_scores, test_scores = learning_curve(estimator, X_train, y_train,cv=10, n_jobs=-1, train_sizes=[1.0, ], scoring="accuracy")
    test_scores = test_scores[0]
    mean, std = test_scores.mean(), test_scores.std()
    return mean, std
def plot(title, xlabel, X, Y, error, ylabel = "Accuracy"):
    plt.xlabel(xlabel)
    plt.title(title)
    plt.grid()
    plt.ylabel(ylabel)
    plt.errorbar(X, Y, error, linestyle='None', marker='o')
```

```python
title = "thay đổi kernel, C = 1"
xlabel = "kernel"
X = []
y = []
error = []
for kernel in tqdm(['linear', 'poly', 'rbf', 'sigmoid']):
    text_clf = SVC(kernel=kernel, C=1.0)
    mean, std = cross_validation(text_clf)
    X.append(kernel)
    y.append(mean)
    error.append(std)
plot(title, xlabel, X, y, error)
plt.savefig('images/svm_change_kernel.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi C, kernel = linear"
xlabel = "C"
X = []
y = []
error = []
for C in tqdm([.1, 1.0, 2.0, 5.0, 10.0]):
    text_clf = SVC(kernel = "linear", C=C)
    mean, std = cross_validation(text_clf)
    X.append(str(C))
    y.append(mean)
    error.append(std)
plot(title, xlabel, X, y, error)
plt.savefig('images/svm_change_C.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi criterion, n_estimators = 50"
xlabel = "criterion"
X = []
y = []
error = []
for val in tqdm(['gini', 'entropy']):
    text_clf = RandomForestClassifier(criterion = val, n_estimators = 50)
    mean, std = cross_validation(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot(title, xlabel, X, y, error)
plt.savefig('images/RF_chang_criterion.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi n_estimators, criterion = gini"
xlabel = "n_estimator"
X = []
y = []
error = []
for val in tqdm([10, 50, 100, 300]):
    text_clf = RandomForestClassifier(criterion = 'gini', n_estimators = val)
    mean, std = cross_validation(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot(title, xlabel, X, y, error)
plt.savefig('images/RF_chang_estimators.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi K"
xlabel = "K"
X = []
y = []
error = []
for val in tqdm([1, 3, 5, 20, 50]):
    text_clf = KNeighborsClassifier(n_neighbors = val)
    mean, std = cross_validation(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot(title, xlabel, X, y, error)
plt.savefig('images/KNN_change_K.png', bbox_inches='tight')
plt.show()
```

```python
svm = SVC(C = 1.0, kernel = 'linear')
svm.fit(X_train, y_train)
```

```python
rf = RandomForestClassifier(n_estimators = 300, criterion = 'gini')
rf.fit(X_train, y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
print("SVM:", accuracy_score(y_test, svm.predict(X_test)))
print("RF:", accuracy_score(y_test, rf.predict(X_test)))
print("KNN:", accuracy_score(y_test, knn.predict(X_test)))
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, svm.predict(X_test))).plot()
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, rf.predict(X_test))).plot()
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, knn.predict(X_test))).plot()
```

```python
data = np.genfromtxt('data/german.data-numeric')
X_data = data[:,:24]
Y_data = data[:,-1]
print(X_data.shape)
print(Y_data.shape)
```

```python
def neg_cost(estimator, X, y):
    y_true = y
    y_pred = estimator.predict(X)
    true_pos = ((y_true==y_pred)&(y_true==1.0))*0.0
    true_ne = ((y_true==y_pred)&(y_true==2.0))*0.0
    false_ne = ((y_true!=y_pred)&(y_true==1.0))*1.0
    false_pos = ((y_true!=y_pred)&(y_true==2.0))*5.0
    return-sum(true_pos + true_ne + false_pos + false_ne)/len(y_true)
```

```python
X_train_, X_test_, y_train_, y_test_ = train_test_split(X_data, Y_data, test_size = 0.2, random_state = 12)
```

```python
def cross_validation_(estimator):
    _, train_scores, test_scores = learning_curve(estimator, X_train, y_train,cv=10, n_jobs=-1, train_sizes=[1.0, ], scoring = neg_cost)
    test_scores = test_scores[0]
    mean, std = test_scores.mean(), test_scores.std()
    return mean, std
def plot_(title, xlabel, X, Y, error, ylabel = "neg_cost"):
    plt.xlabel(xlabel)
    plt.title(title)
    plt.grid()
    plt.ylabel(ylabel)
    plt.errorbar(X, Y, error, linestyle='None', marker='o')
```

```python
title = "thay đổi kernel, C = 1"
xlabel = "kernel"
X = []
y = []
error = []
for val in tqdm(['linear', 'poly', 'rbf', 'sigmoid']):
    text_clf = SVC(kernel = val, C = 1.0)
    mean, std = cross_validation_(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot_(title, xlabel, X, y, error)
plt.savefig('images/SVM_change_kernel_.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi C, kernel = linear"
xlabel = "kernel"
X = []
y = []
error = []
for val in tqdm([.1, 1.0, 3.0, 5.0, 10.0]):
    text_clf = SVC(kernel = "linear", C = val)
    mean, std = cross_validation_(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot_(title, xlabel, X, y, error)
plt.savefig('images/SVM_change_C_.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi criterion, n_estimators = 50"
xlabel = "criterion"
X = []
y = []
error = []
for val in tqdm(['gini', 'entropy']):
    text_clf = RandomForestClassifier(n_estimators = 50, criterion = val) 
    mean, std = cross_validation_(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot_(title, xlabel, X, y, error)
plt.savefig('images/RF_change_criterion.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi estimators, criterion = gini"
xlabel = "n_estimators"
X = []
y = []
error = []
for val in tqdm([5, 10, 50, 100, 300]):
    text_clf = RandomForestClassifier(n_estimators = val, criterion = 'gini') 
    mean, std = cross_validation_(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot_(title, xlabel, X, y, error)
plt.savefig('images/RF_change_estimators.png', bbox_inches='tight')
plt.show()
```

```python
title = "thay đổi K"
xlabel = "K"
X = []
y = []
error = []
for val in tqdm([1, 3, 5, 20, 50]):
    text_clf = KNeighborsClassifier(n_neighbors = val)
    mean, std = cross_validation_(text_clf)
    X.append(str(val))
    y.append(mean)
    error.append(std)
plot_(title, xlabel, X, y, error)
plt.savefig('images/KNN_change_neighbors.png', bbox_inches='tight')
plt.show()
```
